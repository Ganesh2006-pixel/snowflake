```python
import matplotlib.pyplot as plt
```

# CREATE GRAPH

```python
G=nx.Graph()
```

## FACT TABLE

```python
G.add_node("Sales Fact",color="violet")
```

# DIMENSION TABLE

```python
G.add_node("Customer")
G.add_node("Product")
G.add_node("Store")
G.add_node("Date")

```

# NORMALIZED SUB-DIMENSION TABLES

```python
G.add_node("City")
G.add_node("State")
G.add_node("Category")
G.add_node("Subcategory")
```

# Connections from Fact Table


```python
G.add_edge("Sales Fact", "Customer")
G.add_edge("Sales Fact", "Product")
G.add_edge("Sales Fact", "Store")
G.add_edge("Sales Fact", "Date")
```

# Snowflake Normalization Connections


```python
G.add_edge("Customer", "City")
G.add_edge("City", "State")
G.add_edge("Product", "Category")
G.add_edge("Category", "Subcategory")
```

# Node Colors


```python
colors = []
for node in G.nodes():
    if node == "Sales Fact":
        colors.append("red")
    else:
        colors.append("lightblue")
```

# Draw Graph


```python
plt.figure(figsize=(12,8))
pos = nx.spring_layout(G, seed=42)
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=colors,
    node_size=3000,
    font_size=10,
    font_weight='bold',
    edge_color='gray'
)
plt.title("Snowflake Schema Diagram", fontsize=15)
plt.show()
```
